```python
import os
filenames = os.listdir('/home/wsuser/work/Lungcancer/trainset')
```

```python
x_train = train_datagen.flow_from_directory(
    r'/home/wsuser/work/Lungcancer/trainset'
    ,target_size = (64,64),batch_size = 32 ,
    class_mode = 'binary',color_mode="grayscale")
x_test = test_datagen.flow_from_directory(
    r'/home/wsuser/work/Lungcancer/testset',
    target_size = (64,64),batch_size = 32 , 
    class_mode = 'binary',color_mode="grayscale")
```

```python
print(x_train.class_indices)
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32,(3,3), input_shape = (64,64,1),activation = "relu"))
```

```python
model.add(MaxPooling2D(pool_size = (2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 128,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 1,kernel_initializer = "random_uniform",activation = "sigmoid"))
```

```python
model.compile(loss = "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train , steps_per_epoch = 4, 
                    epochs = 100, validation_data = x_test, 
                    validation_steps = 2)
```

```python
model.save("LungCancer.h5")
```
